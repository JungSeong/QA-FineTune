## 1. Custom LLM 클래스 정의

```python
import asyncio
from openai import OpenAI, AsyncOpenAI
from deepeval.models.base_model import DeepEvalBaseLLM


class LocalExaone(DeepEvalBaseLLM):
    """
    vLLM 등 OpenAI-compatible 서버에 연결하는 로컬 LLM 래퍼.

    deepeval 3.8.x 의 DeepEvalBaseLLM 은 내부적으로
      - generate(prompt)  → str
      - a_generate(prompt) → str  (async)
      - get_model_name()  → str
    세 메서드를 요구합니다.
    """

    def __init__(self, model_name: str, base_url: str = "http://localhost:8002/v1"):
        self.model_name = model_name
        self.base_url = base_url
        # 동기 / 비동기 클라이언트 분리 보관
        self._client: OpenAI | None = None
        self._async_client: AsyncOpenAI | None = None

    # ── deepeval 필수 메서드 ─────────────────────────────────────────────

    def load_model(self):
        """동기 클라이언트를 반환합니다 (lazy init)."""
        if self._client is None:
            self._client = OpenAI(api_key="EMPTY", base_url=self.base_url)
        return self._client

    def generate(self, prompt: str, **kwargs) -> str:
        client = self.load_model()
        response = client.chat.completions.create(
            model=self.model_name,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.1,
            **kwargs,
        )
        return response.choices[0].message.content

    async def a_generate(self, prompt: str, **kwargs) -> str:
        """비동기 생성 — AsyncOpenAI 클라이언트를 별도로 사용합니다."""
        if self._async_client is None:
            self._async_client = AsyncOpenAI(api_key="EMPTY", base_url=self.base_url)
        response = await self._async_client.chat.completions.create(
            model=self.model_name,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.1,
            **kwargs,
        )
        return response.choices[0].message.content

    def get_model_name(self) -> str:
        return self.model_name
```

## 2. Custom Embedding 클래스 정의

```python
from sentence_transformers import SentenceTransformer
from deepeval.models.base_model import DeepEvalBaseEmbeddingModel


class LocalEmbedding(DeepEvalBaseEmbeddingModel):
    """
    SentenceTransformer 기반 임베딩 래퍼.

    deepeval 3.8.x 의 DeepEvalBaseEmbeddingModel 은
      - embed_text(text)        → list[float]
      - embed_texts(texts)      → list[list[float]]
      - a_embed_text(text)      → list[float]   (async)
      - a_embed_texts(texts)    → list[list[float]] (async)
      - get_model_name()        → str
    을 요구합니다.

    ※ encode() 결과가 numpy array 이므로 반드시 .tolist() 로 변환합니다.
    """

    def __init__(self, model_name: str):
        self.model_name = model_name
        self._model: SentenceTransformer | None = None

    # ── deepeval 필수 메서드 ─────────────────────────────────────────────

    def load_model(self) -> SentenceTransformer:
        if self._model is None:
            self._model = SentenceTransformer(self.model_name)
        return self._model

    def embed_text(self, text: str) -> list[float]:
        model = self.load_model()
        return model.encode([text], convert_to_numpy=True)[0].tolist()

    def embed_texts(self, texts: list[str]) -> list[list[float]]:
        model = self.load_model()
        return model.encode(texts, convert_to_numpy=True).tolist()

    async def a_embed_text(self, text: str) -> list[float]:
        # SentenceTransformer 는 동기 라이브러리 → executor 에서 실행
        loop = asyncio.get_event_loop()
        return await loop.run_in_executor(None, self.embed_text, text)

    async def a_embed_texts(self, texts: list[str]) -> list[list[float]]:
        loop = asyncio.get_event_loop()
        return await loop.run_in_executor(None, self.embed_texts, texts)

    def get_model_name(self) -> str:
        return self.model_name
```

## 3. 모델 인스턴스 생성 및 전역 임베딩 등록

```python
from deepeval import set_active_embedding_model

# ── 인스턴스 생성 ──────────────────────────────────────────────────────
local_llm = LocalExaone(
    model_name="/models/Exaone-3.5-32B-Instruct",
    base_url="http://localhost:8002/v1",
)

local_embed = LocalEmbedding(
    model_name="dragonkue/snowflake-arctic-embed-l-v2.0-ko"
)

# ── 임베딩 모델을 deepeval 전역으로 등록 ────────────────────────────────
# Synthesizer 가 내부에서 임베딩이 필요할 때 이 모델을 자동으로 사용합니다.
set_active_embedding_model(local_embed)

print("LLM     :", local_llm.get_model_name())
print("Embedding:", local_embed.get_model_name())
print("준비 완료!")
```

## 4. Synthesizer 초기화

```python
from deepeval.synthesizer import Synthesizer

# embedder 파라미터는 set_active_embedding_model() 로 대체했으므로 생략합니다.
synthesizer = Synthesizer(model=local_llm)

print("Synthesizer 초기화 완료")
```

## 5. 골든 데이터셋 생성

> `document_paths` 에 실제 PDF 경로를 넣어 주세요.

```python
# ── 생성할 문서 경로 ───────────────────────────────────────────────────
DOCUMENT_PATHS = ["your_knowledge_base.pdf"]  # ← 실제 경로로 변경
OUTPUT_JSON    = "exaone_dataset.json"

goldens = synthesizer.generate_goldens_from_docs(
    document_paths=DOCUMENT_PATHS,
    max_goldens_per_context=2,
    include_expected_output=True,
)

print(f"생성된 golden 수: {len(goldens)}")
```

## 6. 결과 저장

```python
# deepeval 3.8.x 의 goldens 는 Golden 객체 리스트입니다.
# save_as() / to_dict() 등 버전에 따라 메서드명이 다를 수 있으므로
# 아래처럼 직접 직렬화하는 방법이 가장 안전합니다.
import json

def golden_to_dict(g):
    """Golden 객체를 JSON-직렬화 가능한 dict 로 변환합니다."""
    return {
        "input":           getattr(g, "input",           None),
        "expected_output": getattr(g, "expected_output", None),
        "context":         getattr(g, "context",         None),
        "retrieval_context": getattr(g, "retrieval_context", None),
    }

data = [golden_to_dict(g) for g in goldens]

with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"저장 완료 → {OUTPUT_JSON}")
print("샘플:", json.dumps(data[0], ensure_ascii=False, indent=2) if data else "(데이터 없음)")
```

---
### 주요 수정 사항 요약

| 항목 | 원본 코드 문제 | 수정 내용 |
|---|---|---|
| `a_generate` | `return self.generate(prompt)` — 동기 함수를 async에서 직접 호출하면 이벤트 루프가 블로킹됨 | `AsyncOpenAI` 클라이언트로 진짜 비동기 처리 |
| `a_embed_text/texts` | `SentenceTransformer`는 동기 라이브러리인데 async 함수에서 바로 호출 | `loop.run_in_executor()` 로 스레드풀에서 실행 |
| `embed_text` 반환값 | `[0].tolist()` — 단일 벡터에서 첫 요소만 반환될 위험 | `convert_to_numpy=True` + 명시적 슬라이싱 보강 |
| `load_model` | 매 호출마다 새 클라이언트/모델 생성 | Lazy init 패턴으로 한 번만 생성 |
| 결과 저장 | `goldens.save_as_json()` — 버전에 따라 메서드 없음 | `json.dump` 로 직접 직렬화 (안전) |
| import 중복 | `from deepeval.models.base_model import ...` 두 번 선언 | 셀 분리 및 중복 제거 |
