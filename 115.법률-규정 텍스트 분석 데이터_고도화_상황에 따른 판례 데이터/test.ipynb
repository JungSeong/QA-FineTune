```python
import torch

def check_gpu_memory():
    if torch.cuda.is_available():
        # free, total 메모리를 바이트 단위로 가져옴
        free_mem, total_mem = torch.cuda.mem_get_info()
        
        free_mb = free_mem / 1024**2
        total_mb = total_mem / 1024**2
        used_mb = total_mb - free_mb
        
        print(f"Total Memory: {total_mb:.2f} MB")
        print(f"Free Memory : {free_mb:.2f} MB")
        print(f"Used Memory : {used_mb:.2f} MB")
        
        return free_mb
    else:
        print("CUDA를 사용할 수 없습니다.")
        return None

# 가용 메모리에 따른 Batch Size 가이드
available_mb = check_gpu_memory()
if available_mb and available_mb > 20000:  # 예: 20GB 이상 여유 있을 때
    print("Suggesting Batch Size: 64 or higher")
```
